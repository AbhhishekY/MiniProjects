```
red=0
blue=0
green=0
no_of_pixels=rows*cols

for i in range(rows):
  for j in range(cols):
    k=img[i,j]
    red+=k[0]
    green+=k[1]
    blue+=k[2]

red=red/no_of_pixels
blue=blue/no_of_pixels
green=green/no_of_pixels
```
